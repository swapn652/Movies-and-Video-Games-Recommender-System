```python
my_data['movie_title'] = my_data['movie_title'].replace("&#39;", "'")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
my_data.dropna(subset=['combined_information'], inplace=True)
```

```python
vectors = cv.fit_transform(my_data['combined_information']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = my_data[my_data['movie_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:10]
    
    for i in movies_list:
        print(my_data.iloc[i[0]].movie_title)
```
